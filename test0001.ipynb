```python
import os
#
# matching dictionaries
MDA = ["a","artifacts","audio"]
MDB = ["b","blue","bike","blocky", "blockiness"]
MDC = ["c","captions","can't","cannot","cannot","cell", "couldn't", "couldn't hear","couldn't watch"]
MDD = ["d","distorted","don't play","do not","did not","didn't play"]
MDE = ["e","error"]
MDF = ["f", "fail","failed","froze","frozen"]
MDG = ["g","garbage"]
MDH = ["h", "hear"]
MDI = ["i"]
MDJ = ["j", "jpeg"]
MDK = ["k"]
MDL = ["l", "lag", "low quality"]
MDM = ["m","macroblock","mpeg", "mosquitoes"]
MDN = ["n", "noise"]
MDO = ["o", "of sync","out"]
MDP = ["p","paused","pausing", "picture stopped", "picture froze","picture frozen"]
MDQ = ["q","quality"]
MDR = ["r","refund"]
MDS = ["s","slow","stutter","stuttered","stuttering","sync", "sync problem"]
MDT = ["t","too loud","too quiet"]
MDU = ["u"]
MDV = ["v", "video paused", "video stuttered"]
MDW = ["w", "was distorted", "wasn't playing"]
MDX = ["x"]
MDY = ["y"]
MDZ = ["z"]
INK = 1
CEND = '\33[0m' ## normal
CBOLD ='\33[1m' ## bold
CITALIC   ='\33[3m'
CBLINK    ='\33[5m'
CBLINK2   ='\33[6m'
CSELECTED ='\33[7m'
CBLACK    ='\33[30m'
CRED      ='\33[31m'
CGREEN    ='\33[32m'
CYELLOW   ='\33[33m'
CBLUE     ='\33[34m'
CVIOLET   ='\33[35m'
TEAL      ='\33[36m'
CWHITE    ='\33[37m'
MDLIST = [MDA,MDB,MDC,MDD,MDE,MDF,MDG,MDH,MDI,MDJ,MDK,MDL,MDM,MDN,MDO,MDP,MDQ,MDR,MDS,MDT,MDU,MDV,MDW,MDX,MDY,MDZ]
MATCHSCORE=0

print ("\n\n")
#
def checkword (myword) :
    #print("myword is",myword)
    if myword == "" :
        return
    firstletter = myword[0]
    myword = myword.replace('.', '')
    global MATCHSCORE
    foo = len(MDLIST)
    for i in range (0,foo):
        currentlist = MDLIST[i]
        #print("current list is ",currentlist)
        currentletter = currentlist[0]
        if (currentletter == firstletter) :
            #print("checking letter ", currentletter, " ...")
            ## do stuff
            length = len(currentlist)
            for j in range (1,length) :
                if (currentlist[j] == myword) :
                    print("  ->",currentlist[j],"")
                    MATCHSCORE += INK
#
 #INPUTTEXT = """This is my short review. The video stuttered and the audio was distorted.
 #I couldn't watch the whole thing due to the jpeg artifacts and bad picture.
 #I couldn't hear the sound properly.  Please fix it."""
#
#
  ##------GET USER PASTED INPUT-------


print(" Please Paste the Review text now:\n"
      " To end recording Press Ctrl+d on Mac/Linux or Crtl+z on Windows. \n\n")
INPUTTEXT = []
try:
    while True:
        INPUTTEXT.append(input())
except EOFError:
    pass
INPUTTEXT = "\n".join(INPUTTEXT)
print("\n\n")
ilen = len(INPUTTEXT)
if (ilen <2 ):
	print("no valid text on clipboard...exiting.")
	exit()

for line in INPUTTEXT.splitlines() :
    currline = line.rstrip('\n') # .rstrip('\n') removes the line break
    words = currline.split(' ')
    wlen = len(words)
    if (wlen % 2 != 0):
        words.append("null")

    for x, y in zip(words, words[1:]):
        myphrz = (x+" "+y).lower()
        if (x != '\n'):
            INK = 1; checkword(x.lower())
            INK = 3; checkword(myphrz.lower())
            INK = 1;

## ---  recommendation ---
print(CEND,"\n\n Keyword matches: ", MATCHSCORE,"\n")
if (MATCHSCORE >= 5):
	print (CRED,CBOLD,"HIGH",CEND,"probability of Compromized Customer Playback Experience. Recommend further investigation ASAP")

elif (MATCHSCORE > 0):
	print (CYELLOW,"MEDIUM",CEND, "probability of Compromized Customer Playback Experience. Recommend further investigation.")

elif (MATCHSCORE == 0):
	print(CEND,"No indications of compromized playback.")
##
print("\ndone.")
```
